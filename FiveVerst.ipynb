```python
import urllib.request
import re
import json

with urllib.request.urlopen('https://5verst.ru/events/') as response:
    encoding = response.info().get_param('charset', 'utf8')
    htmlStr = response.read().decode(encoding)

r = re.compile('name:\s*"(?P<name>.*?)".*coord:\s*"(?P<coord>.*?)".*url:\s*"(?P<url>.*?)"')
places = [m.groupdict() for m in r.finditer(htmlStr)]

print(len(places), ' places found:')
print()

jsonStr = json.dumps(places, ensure_ascii=False)
formattedJsonStr = jsonStr.replace('}, {', '},\n  {')

print(formattedJsonStr)
```
